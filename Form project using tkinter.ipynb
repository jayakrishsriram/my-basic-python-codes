```python
import tkinter as tk
from tkinter import messagebox

def button_click():
    print("Name of the Student:", entry.get().title())
    print("Gender:", radio_var.get().title())
    print("Department:", radio_varb.get().title())
    print("Section:", clicked.get().title())
    print("Location:", entry1.get().title())
    print("Your option is/are:")
    if(c1.get()==1):  
        print("Python",end=",")
    if(c2.get()==1):  
        print("Java",end=",")
    if(c3.get()==1):  
        print("Ruby")
    mes=messagebox.showinfo("Submitted","Your information is successfully submitted")

root = tk.Tk()
root.title("My Form")

#menu to exit
menu = tk.Menu(root)
menu.add_command(label="Exit", command=root.destroy)
root.config(menu=menu)
root.configure(background="sky blue")

#Entry details
label = tk.Label(root, text="What is your name?")
label.grid(row=0)
entry = tk.Entry(root)
entry.grid(row=0,column=1)

labele = tk.Label(root, text="Location:")
labele.grid(row=6)
entry1 = tk.Entry(root)
entry1.grid(row=6,column=1)

#radio button for gender
radio_var = tk.StringVar()
label = tk.Label(root, text="Your gender:")
label.grid(row=1)

radio1 = tk.Radiobutton(root, text="Male", variable=radio_var, value="Male")
radio1.grid(row=1,column=1)

radio2 = tk.Radiobutton(root, text="Female", variable=radio_var, value="Female")
radio2.grid(row=1,column=2)

radio3 = tk.Radiobutton(root, text="Other", variable=radio_var, value="Other")
radio3.grid(row=1,column=3)

#radio button for Department
radio_varb = tk.StringVar()
labelb = tk.Label(root, text="Your Department")
labelb.grid(row=2)

radiob1 = tk.Radiobutton(root, text="CSE", variable=radio_varb, value="CSE")
radiob1.grid(row=2,column=1)

radiob2 = tk.Radiobutton(root, text="ECE", variable=radio_varb, value="ECE")
radiob2.grid(row=2,column=2)

radiob3 = tk.Radiobutton(root, text="CIVIL", variable=radio_varb, value="CIVIL")
radiob3.grid(row=2,column=3)

#Dropdown menu
labeld = tk.Label(root, text="Your Section")
labeld.grid(row=3)

options = ["Select","A","B","C","D"]
# datatype of menu text
clicked = tk.StringVar()
  
# initial menu text
clicked.set( "Select" )
  
# Create Dropdown menu
drop = tk.OptionMenu( root , clicked , *options )
drop.grid(row=3,column=2)

#checkbox
c1 = tk.IntVar()
c2 = tk.IntVar()
c3 = tk.IntVar()
label = tk.Label(root, text="Known programming language:")
label.grid(row=5)
checkbox1 = tk.Checkbutton(root, text="Python", variable=c1)
checkbox1.grid(row=5,column=1)
checkbox2 = tk.Checkbutton(root, text="Java", variable=c2)
checkbox2.grid(row=5,column=2)
checkbox3 = tk.Checkbutton(root, text="Ruby", variable=c3)
checkbox3.grid(row=5,column=3)

button = tk.Button(root, text="Submit", command=button_click)


button.grid(row=10,column=10)
root.mainloop()
```
